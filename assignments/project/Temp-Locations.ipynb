```python
import pandas as pd
import numpy as np
import en_core_web_sm
import nltk
nlp = en_core_web_sm.load()
```

```python
df = pd.read_csv("output_combined.csv")
df.head()
```

```python
world = pd.read_csv('https://raw.githubusercontent.com/datasets/world-cities/master/data/world-cities.csv')
cities = list(world['name'].str.lower().unique())
countries = list(world['country'].str.lower().unique())
```

```python
def extract_loc(row):
    tweet = row['text']
    token = nltk.word_tokenize(tweet)
    tags = nltk.pos_tag(token)
    nouns = list(map(lambda y: y[0].lower(), filter(lambda x: x[1] == 'NNP' or x[1] == 'NN', tags)))
    country = [value for value in nouns if value in countries]
    city = [value for value in nouns if value in cities]
    intersect = city + country
    intersect = list(filter(lambda x: x != 'holiday', intersect))
    return list(map(lambda x: x.title(), intersect)) if len(intersect) > 0 else np.nan
```

```python
temp = df.copy()
temp['locations'] = temp.apply(extract_loc, axis=1)
mask = temp['locations'].notnull()
print(len(temp[mask]))
temp[mask].locations[0:10]
```
