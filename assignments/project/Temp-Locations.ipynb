```python
import pandas as pd
import numpy as np
import en_core_web_sm
import nltk
nlp = en_core_web_sm.load()
```

```python
df = pd.read_csv("data/output_combined.csv")
df
```

```python
world = pd.read_csv('https://raw.githubusercontent.com/datasets/world-cities/master/data/world-cities.csv')
cities = list(world['name'].str.lower().unique())
countries = list(world['country'].str.lower().unique())
```

```python
def extract_locations(row, t):
    tweet = row['text']
    tweet = tweet.replace('#', '')
    token = nltk.word_tokenize(tweet)
    tags = nltk.pos_tag(token)
    nouns = list(map(lambda y: y[0].lower(), filter(lambda x: x[1] == 'NNP' or x[1] == 'NN', tags)))
    
    l = []
    if (t == 'country'):
        l = [value for value in nouns if value in countries]
    elif (t == 'city'):
        l = [value for value in nouns if value in cities]
        l = list(filter(lambda x: x != 'holiday', l))
        
    return list(map(lambda x: x.title(), l)) if len(l) > 0 else np.nan
```

```python
temp = df.copy()
temp['cities'] = temp.apply(lambda x: extract_locations(x, 'city'), axis=1)
temp['countries'] = temp.apply(lambda x: extract_locations(x, 'country'), axis=1)
temp.head()
```

```python
a = temp.dropna(subset=['cities', 'countries'], how="all")
a.head()
```

```python
print('found {} rows with locations'.format(len(a)))
```

```python
def mapcountries(row):
    cities = row['cities']
    countries = []
    if (type(cities) == list):
        for c in cities:
            country = list(world[world['name'] == c].country)
            if (len(c) > 0):
                countries = countries + country
        if (len(countries) > 0):
            row['countries'] = countries
    return row

temp = temp.apply(mapcountries, axis=1)
```
